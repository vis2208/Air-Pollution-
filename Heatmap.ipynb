```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/vis2208/Air-Pollution-/main/Processed_data.csv")
df.head(5)
```

```python
import seaborn as sns
```

```python
df1 = df[["PM2.5","PM10","NO2","SO2","CO"]]
df1.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

size = len(df1.columns)
f,ax = plt.subplots(figsize=(size, size))
sns.heatmap(df1.corr().abs(), annot=True, linewidths=.5, fmt= '.2f',ax=ax)
```

```python
df2 = df[["PM2.5","RH","WS","WD","SR","AT"]]
df2.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

size = len(df2.columns)
f,ax = plt.subplots(figsize=(size, size))
sns.heatmap(df2.corr().abs(), annot=True, linewidths=.5, fmt= '.2f',ax=ax)
```
