# Create DataFrame

```python
df = pd.concat([df1,df2,df3,df4,df5])
df.head(5)
```

```python
df.drop(['To Date'], axis = 1, inplace = True ) #dropped column "To Date"
```

# Change Date format


```python
df['From Date'] = df['From Date'].str.replace('/','-')
```

```python
df['From Date'] = pd.to_datetime(df['From Date'], format = '%d-%m-%Y %H:%M')
```

```python
df['From Date'] = df['From Date'].dt.strftime('%d/%m/%Y')
```

```python
df.head(30)
```

# Rename Column Name and Set index column

```python
df.rename(columns = {'From Date':'Date'},inplace = True)
```

```python
df.set_index('Date', inplace = True)
```

# Remove New Year & Diwali data 


```python
df.drop(['01/01/2018','02/01/2018','01/01/2019','02/01/2019','01/01/2020','02/01/2020','08/11/2018','09/11/2018','28/10/2019','29/10/2019'],inplace=True)
```

```python
df.replace("None",np.NaN , inplace = True) # replaced "None" by NaN
```

```python
df.isnull().sum()  #number of NaN in each column
```

```python
df.dropna(how = "all", inplace = True) #dropped those rows in which all columns are NaN
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['PM2.5'].dtype,df['PM10'].dtype,df['NO2'].dtype,df['SO2'].dtype,df['CO'].dtype
```

```python
df['PM2.5'] = df['PM2.5'].astype('float')
df['PM10'] = df['PM10'].astype('float')
df['NO2'] = df['NO2'].astype('float')
df['SO2'] = df['SO2'].astype('float')
df['CO'] = df['CO'].astype('float')
df['RH'] = df['RH'].astype('float')
df['WS'] = df['WS'].astype('float')
df['WD'] = df['WD'].astype('float')
df['SR'] = df['SR'].astype('float')
df['AT'] = df['AT'].astype('float')
```

```python
df['PM2.5'].dtype,df['PM10'].dtype,df['NO2'].dtype,df['SO2'].dtype,df['CO'].dtype
```

```python
boxplot = df.boxplot(column=["PM2.5","PM10","NO2","SO2","CO"],grid = False, fontsize = 15, rot = 90)  #box-plot
```

```python
cols = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'RH', 'WS', 'WD', 'SR', 'AT'] # one or more

Q1 = df[cols].quantile(0.25)
# print(Q1)
Q3 = df[cols].quantile(0.75)
# print(Q3)
IQR = Q3 - Q1
# print(IQR)
df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
df.head()
```

```python
df.shape
```

```python
average = df.mean() #mean 
average
```

```python
df = df.replace(np.NaN,average) #replaced NaN by mean value of corresponding column
df.head(5)
```

```python
df.info() #info of final processed data
```

```python
df.to_csv("Processed_data.csv")  #write processed data
```
